```python
import pandas as pd 
import numpy as np 
# import matplotlib.pyplot as plt
import re
```

```python
all_df = pd.read_excel('보안정보.xlsx')
all_df.head()
print(all_df.shape)
```

```python
address = all_df['주소']
```

```python
# 주소에서 괄호제거를 위한 함수 
def clean_text(data):
    data = re.sub('[\(\)]','',data)
    return data
```

```python
#괄호제거한 값들 새롭게 할당 
new_address=[]
for i in address:
    i = clean_text(str(i))
    new_address.append(i)
        
```

```python
#읍,면,동으로 끝나는 자연어를 추출 해 리스트 형태로 반환하고 새로운 리스트 생성 (dong)
dong =[]
q=re.compile('.{2,3}면|.{2,3}읍|.{2,3}동')
for i in new_address:
        m =q.findall(str(i)) # 여기서 해당 문자열이 없으면 None 반환함 
        
        if m is None:
            dong.append(str(i)) # None 값을 처리하는 함수 
        else:
#             group = m.group()
            dong.append(m) 
        # None 타입 잇는 경우 어떻게 할래 
```

```python
# 숫자 제거 
r= re.compile('^[0-9]')
for i in dong:
    if len(i)==1:
        i[0] = re.sub('[0-9]','',i[0])
    elif len(i)==2 :
        i[0] = re.sub('[0-9]','',i[0])
        i[1] = re.sub('[0-9]','',i[1])
    else:
        i
    
```

```python
# 공백 제거 
r= re.compile('^[0-9]')
for i in dong:
    if len(i)==1:
        i[0] = re.sub('\s','',i[0])
    elif len(i)==2 :
        i[0] = re.sub('\s','',i[0])
        i[1] = re.sub('\s','',i[1])
    else:
        i
    
```

```python
real_dong =[]
for i in dong:
    if len(i)==1:
        real_dong.append(i[0])
    elif len(i)==2:
        if re.search('읍$|면$',str(i[0])) is None: # 읍또는 면으로 끝나지 않는 경우 1번째 리스트 값 반영 
            real_dong.append(i[1])
        else:
            real_dong.append(i[0]) # 읍,또는 면으로 끝나는 경우 리스트 반영 
    else:
        real_dong.append('먹깨비') # 빈스르트는 len이 0이다 따라서 else로 처리 
        
```

```python
com_lst = []
for i in real_dong:
    i = re.sub('[\-]','',str(i)) # 수동 -제거 
    if len(i) == 4: #4개 나왓던 것들 뒤에서 3개 추출 
        com_lst.append(i[-3:])
    else:
        com_lst.append(i)

com_lst # 동 추출 완료 

```

```python
## 근데 예를들어 수곡1동 이런 것들 숫자 제거할때 1까지 다 제거해버림 
## 살리는 방법 탐구 
```

```python

all_df['dong'] = com_lst
```
